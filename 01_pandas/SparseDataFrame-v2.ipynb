```python
#https://towardsdatascience.com/hacking-scikit-learns-vectorizers-9ef26a7170af
import pandas as pd
import numpy as np
df = pd.DataFrame(np.arange(0,100).reshape(10,10))
df[4:7]=0
df
```

```python
import numpy as np
df.to_sparse().density
```

```python
df.to_dense().to_sparse().density
```

```python
#稀疏矩阵转稀疏矩阵,不会触发fill_value重新设置
df.to_dense().to_sparse().to_sparse(fill_value=0).density
```

```python
#只有真正除非稀疏矩阵转换fill_value才会起效
df.to_dense().to_sparse().to_dense().to_sparse(fill_value=0).density
```

```python
#设置了fill_value=0后, nan也会被当成正常值处理
df[4:7]=np.nan
df.to_dense().to_sparse().to_dense().to_sparse(fill_value=0).density
```
