# 有多个Index时,如何利用部分index定位数据

```python
#https://www.somebits.com/~nelson/pandas-multiindex-slice-demo.html
import pandas, io
data = io.StringIO('''Fruit,Color,Count,Price
Apple,Red,3,$1.29
Apple,Green,9,$0.99
Pear,Red,25,$2.59
Pear,Green,26,$2.79
Lime,Green,99,$0.39
''')
df_unindexed = pandas.read_csv(data)
df_unindexed
```

```python
df = df_unindexed.set_index(['Fruit', 'Color'])
df
```

```python
df.loc['Apple', :]   
```

```python
df.loc[('Apple', ), :] 
```

```python
df.xs('Red', level='Color')
```
