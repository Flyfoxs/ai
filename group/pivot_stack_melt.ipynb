```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.arange(0, 12).reshape((4,3)), columns=['a','b','c'])

df.b =df.b%2
df1 = df.copy()
df2 = df.copy()
df1['who'] = 'Mike'
df2['who'] = 'Bob'

df = pd.concat([df1, df2])
df.a = np.arange(0,8)
df = df.reset_index(drop=True)
```

```python
df.pivot(index='a', columns='b', values='c')
```

```python
#用Unstack来实现 Pivot
s = df[['a','b','c']]
s = s.set_index(['a','b'])
s = s.unstack(level='b')
s
```

```python
##效果等价于Pivot
xx = df.stack()
print(type(xx))
xx
```

```python
print(xx[(0,'c')])
print(xx[(0,'who')])
```
