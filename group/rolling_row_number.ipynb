```python
import pandas as pd
import numpy as np


df1 = pd.DataFrame({'A':'a',
             'B':range(0,3)})

df2 = pd.DataFrame({'A':'b',
             'B':np.random.randn(5)})

df = pd.concat([df1, df2])

df
```

## 实现数据库的row_number, 分组排序

```python
df = df.sort_values(['A','B'], ascending=True)
gp = df.groupby('A')

######下面2行不能互换位置######
df['max'] = gp.cummax()
df['row_number'] = gp.cumcount()

df
```

## 每个分组取出一条数据(第n条)

```python
df.groupby('A').nth(0)
```

## 分组后, 使用滑动窗口进行聚合

```python
df = df.reset_index(drop=True)
roll = df.groupby('A')['B'].rolling(2).sum()
roll = roll.reset_index(level=0, drop=True)
df['roll'] = roll
df
```
