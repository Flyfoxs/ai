```python
import pandas as pd
import numpy as np
df_1 = pd.DataFrame(
    {'month':np.arange(0, 6, 1),
     'value':np.random.randn(6) ,
     'who':'Bob'
    }    
)

df_2 = pd.DataFrame(
    {'month':np.arange(0, 6, 1),
     'value':np.random.randn(6) ,
     'who':'Alice'
    }    
)

df = pd.concat([df_1, df_2])

df = df.reset_index(drop=True)
```

```python
#这种做法如果月份数据有丢失,会自然的比较连续的2条数据
df['previous'] = df.groupby('who')['value'].shift(1)
df['percentage'] = (df.value-df.previous )/df.previous
df
```
