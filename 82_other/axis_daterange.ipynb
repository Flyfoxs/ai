```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(24*60*100).reshape(100, 24*60 ), 
                  index = pd.date_range('2018-01-01',freq='D',periods= 100) )

df.head()
```

```python
df['max'] = df.apply(lambda row: max(row), axis=1)

df['min_sn'] = df.apply(lambda row: row.idxmax() , axis=1)

df[['max', 'min_sn']].head()
```
