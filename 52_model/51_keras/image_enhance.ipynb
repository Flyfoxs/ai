```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
img = plt.imread('../98_sample/1.jpg')


from keras.preprocessing.image import ImageDataGenerator


def transfer_img(datagen):

    img = plt.imread('../98_sample/1.jpg')
    img = np.expand_dims (img, axis=0)

    grid_size = (2, 4)
    fig = plt.figure(figsize=(12, 6)) 

    for i in range(grid_size[0]):
      for j in range(grid_size[1]):
        if i==0 and j==0:
            X_batch = img
        else:
            X_batch = datagen.flow(img, None, batch_size=1)[0]
        axes = plt.subplot2grid(grid_size, (i, j), rowspan=1, colspan=1)
        axes.get_xaxis().set_visible(False)
        axes.get_yaxis().set_visible(False)

        plt.imshow(X_batch[0])
        #plt.title('%d, %d, %d, %d' % tuple(params), fontsize = 'small')

    plt.ticklabel_format()
    plt.show()
    

```

```python
datagen = ImageDataGenerator(width_shift_range=0.1,
                                     height_shift_range=0.1,
                                     zoom_range=0.1,
                             #rescale=1. / 255,
                            )
transfer_img(datagen)
```

```python
 datagen = ImageDataGenerator(
            # 图片随机水平偏移的幅度❎
            width_shift_range=0.1,
            # 图片随机垂直偏移的幅度,❎
            height_shift_range=0.1,

            # 随机放大
            zoom_range=0.4,
            # 剪切强度
            shear_range=0.4,
            # 图片随机翻转的角度
            rotation_range=30,
            # 随机水平翻转
            horizontal_flip=True,
            # # 执行其他处理前乘到整个图像上
            rescale=1. / 255,
        )
        
transfer_img(datagen)
```

```python
 datagen = ImageDataGenerator(
            # 图片随机水平偏移的幅度❎
            width_shift_range=0.1,
            # 图片随机垂直偏移的幅度,❎
            height_shift_range=0.1,

            # 随机放大
            zoom_range=0.1,
            # 剪切强度
            shear_range=0.1,
            # 图片随机翻转的角度
            rotation_range=10,
            # 随机水平翻转
            horizontal_flip=True,
            # # 执行其他处理前乘到整个图像上
            #rescale=1. / 255,
        )
        
transfer_img(datagen)
```
