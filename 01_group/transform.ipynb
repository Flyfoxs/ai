```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.arange(0, 14).reshape(7,2),columns =['a','b'] )

df.a =  df.a %3
df['who'] = 'Bob'
df.iloc[6,1]=np.nan
df
```

```python
### Transform妙用,插入聚合结果到原数据结果时,不用通过Join/Merge来实现
```

```python
df['b_mean'] = df.sort_values('a').groupby('a')['b'].transform(np.mean)

df.sort_values('a')
```

```python
#Fillna不需要用全局均值,可以考虑用分组均值,这样准确度会更高些
df['b'] = df.groupby('a')['b'].transform(lambda x: x.fillna(x.mean()))
df
```
