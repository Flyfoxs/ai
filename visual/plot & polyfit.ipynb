## 一个Plot画出2条直线

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x = range(0, 10)
plt.plot(x, x)

y = range(4, 10)
plt.plot(y, list(map(lambda val: val*2, y ) ) )
```

## 多项式自动拟合

```python
x = list(np.arange(0,4,0.4))

y = list(map(lambda val: val**3*3 + np.random.random()*20 , x) )

plt.scatter(x, y)

w = np.polyfit (x, y, 4)

fn = np.poly1d(w)

print(w)
print(fn)

plt.plot(x, fn(x))
```
