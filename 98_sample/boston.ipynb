```python
#http://scikit-learn.org/stable/datasets/index.html

from sklearn import datasets
import pandas as pd
import numpy as np
data = datasets.load_boston()
df = pd.DataFrame( data.data , columns=data.feature_names)
df['target'] = data.target
df.head()
```

```python
tmp = df.groupby(['ZN','B']).agg({'RAD':['count', 'sum'], 'TAX':'nunique'})
tmp.head()
```

```python
tmp2 = tmp.reset_index()
tmp3 = tmp2.pivot(index='ZN', columns='B')
tmp3.head()
```

```python
tmp3.columns = ['_'.join( [ str(value) for value in item]) for item in  tmp3.columns]
tmp3.head()
```
