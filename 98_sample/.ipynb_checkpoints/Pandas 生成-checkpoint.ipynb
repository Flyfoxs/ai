```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.arange(0, 6).reshape(3,2),columns =[1,'a'] , index=['a','b','a'])
df['who'] = 'Alice'
df.loc['a','who'] =  'Bob'
df
```

```python

import pandas as pd
import numpy as np
df = pd.DataFrame(np.arange(0, 14).reshape(7,2),columns =['a','b'] )

df.a =  df.a %3
df['who'] = 'Bob'
df.iloc[6,1]=np.nan
df
```

```python
import pandas as pd
import numpy as np
df_1 = pd.DataFrame(
    {'month':np.arange(0, 6, 1),
     'value':np.random.randn(6) ,
     'who':'Bob'
    }    
)

df_2 = pd.DataFrame(
    {'month':np.arange(0, 6, 1),
     'value':np.random.randn(6) ,
     'who':'Alice'
    }    
)

df = pd.concat([df_1, df_2])

df = df.reset_index(drop=True)
df
```
